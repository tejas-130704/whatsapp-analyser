```python
import re
import pandas as pd
```

```python
f = open("WhatsApp Chat with Tighe.txt","r",encoding="utf-8")
```

```python
isTime12Format=False
data= f.read()
```

```python
print(data)
```

```python
if re.search(r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s',data):
    #pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s'
    isTime12Format=True
    pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s'
else:
    pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
    
```

```python

messages = re.split(pattern, data.strip())[1:]
print(messages)
```

```python
dates = re.findall(pattern,data)
```

```python
dates
```

```python
cleaned_dates = [date.replace('\u202f', ' ').strip(' - ') for date in dates]
```

```python
cleaned_dates
```

```python
df=pd.DataFrame({'Messages':messages,"Date":cleaned_dates})
```

```python
if isTime12Format:
    df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%y, %I:%M %p")
else:
    df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y, %H:%M")
```

```python
df.sample(5)
```

```python
df["Messages"]=df["Messages"].apply(lambda x:x.replace("\n",""))
```

```python
df.sample(5)
```

```python
name_pattern = r'^([^:]+):'

#df["Name"]=df["Messages"].apply(lambda x: re.match(name_pattern, x).group(1) if re.match(name_pattern, x) else "Notification")
```

```python
df.sample(5)
```

```python
users=[]
messages=[]
for msg in df["Messages"]:
    entry=re.split(r'^([^:]+):',msg)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append("Notification")
        messages.append(entry[0])

df["User"]=users
df["Messages"]=messages
        
```
